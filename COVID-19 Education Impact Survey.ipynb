# Exploratory Data Analysis:
# WHAT TYPE OF PEOPLE IS THE MOST IMPACTED EDUCATIONALLY BY COVID_19?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cv_edu_impact = pd.read_csv('../data/open_one_time_covid_education_impact.csv')
cv_edu_impact.head()
```

```python
sns.countplot(y='age', data=cv_edu_impact).set(title='Age distribution')
```

```python
sns.countplot(y='education', data=cv_edu_impact).set(title='Educational background distribution')
```

## As seen above, people who completed some degrees or completed secondary school, high school are the most impacted by COVID-19. Plus, people around these ages are also distributing most in the dataset, which makes sense


```python
sns.countplot(y='geography', data=cv_edu_impact).set(title='Geography distribution')
```

## People who live in city center or metropolitan area distribute most in the dataset

```python
sns.countplot(y='gender', data=cv_edu_impact).set(title='Gender distribution')
```

```python
labels = cv_edu_impact.are_children_being_teached_by_unqualified_people.unique()
sizes = cv_edu_impact.are_children_being_teached_by_unqualified_people.value_counts().unique()
colors = ['yellowgreen', 'red']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)
plt.title("Are children being teached by unqualified people?")
```
